# Verificações de Validação
Notebook utilitário para validar rapidamente as contagens de registros nas camadas Bronze, Silver e Gold de um determinado ambiente. Não faz parte do workflow orquestrado — executar manualmente após uma execução do pipeline.

```python
dbutils.widgets.text(name="env", defaultValue="", label="Enter the environment in lower case")
env = dbutils.widgets.get("env")
```

### Contagens da camada Bronze

```python
df_roads = spark.sql(f"SELECT COUNT(*) AS road_count FROM `{env}_catalog`.`bronze`.`raw_roads`")
print("raw_roads:", df_roads.collect()[0]["road_count"])

df_traffic = spark.sql(f"SELECT COUNT(*) AS traffic_count FROM `{env}_catalog`.`bronze`.`raw_traffic`")
print("raw_traffic:", df_traffic.collect()[0]["traffic_count"])
```

### Contagens da camada Silver

```python
df_roads = spark.sql(f"SELECT COUNT(*) AS road_count FROM `{env}_catalog`.`silver`.`silver_roads`")
print("silver_roads:", df_roads.collect()[0]["road_count"])

df_traffic = spark.sql(f"SELECT COUNT(*) AS traffic_count FROM `{env}_catalog`.`silver`.`silver_traffic`")
print("silver_traffic:", df_traffic.collect()[0]["traffic_count"])
```

### Contagens da camada Gold

```python
df_roads = spark.sql(f"SELECT COUNT(*) AS road_count FROM `{env}_catalog`.`gold`.`gold_roads`")
print("gold_roads:", df_roads.collect()[0]["road_count"])

df_traffic = spark.sql(f"SELECT COUNT(*) AS traffic_count FROM `{env}_catalog`.`gold`.`gold_traffic`")
print("gold_traffic:", df_traffic.collect()[0]["traffic_count"])

df_traffic_summary = spark.sql(f"SELECT COUNT(*) AS row_count FROM `{env}_catalog`.`gold`.`traffic_summary`")
print("traffic_summary:", df_traffic_summary.collect()[0]["row_count"])

df_traffic_region = spark.sql(f"SELECT COUNT(*) AS row_count FROM `{env}_catalog`.`gold`.`traffic_by_region`")
print("traffic_by_region:", df_traffic_region.collect()[0]["row_count"])

df_road_usage = spark.sql(f"SELECT COUNT(*) AS row_count FROM `{env}_catalog`.`gold`.`road_usage_statistics`")
print("road_usage_statistics:", df_road_usage.collect()[0]["row_count"])

df_ev_trend = spark.sql(f"SELECT COUNT(*) AS row_count FROM `{env}_catalog`.`gold`.`ev_adoption_trend`")
print("ev_adoption_trend:", df_ev_trend.collect()[0]["row_count"])
```
